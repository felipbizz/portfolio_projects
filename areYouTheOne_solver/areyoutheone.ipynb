```python
import itertools
import pickle
```

```python
# homens e mulheres em ordem alfabética
guys = ["Anthony", "Clinton", "Dimitri", "Ethan", "Joe", "Kareem", "Keith", "Malcom", "Michael", "Shad", "Tyler"]
girls = ["Alexis", "Alivia", "Audrey", "Diandra", "Geles", "Jada", "Keyana", "Nicole", "Nurys", "Uche", "Zoe"]

# (palpites, numero de matchs)
# Lista de mulheres que fizeram casal com o respectivo homem. exemplo: week1, "Francesca" foi com "Asaf"
week1 = (["Geles", "Uche", "Diandra", "Jada", "Zoe", "Alivia", "Alexis", "Nurys", "Keyana", "Audrey", "Nicole"], 3)
# week2 = (["Diandra", "Uche", "Nicole", "Jada", "Audrey", "Alivia", "Alexis", "Nurys", "Keyana", "Geles", "Zoe"], 1)
# week3 = (["Jada", "Uche", "Nurys", "Alexis", "Zoe", "Alivia", "Diandra", "Geles", "Audrey", "Keyana", "Nicole"], 2)
# week4 = (["Keyana", "Uche", "Alexis", "Nicole", "Zoe", "Diandra", "Nurys", "Alivia", "Geles", "Audrey", "Jada"], 3)
# week5 = (["Nicole", "Jada", "Uche", "Geles", "Zoe", "Alivia", "Alexis", "Diandra", "Nurys", "Audrey", "Keyana"], 1)
# week6 = (["Keyana", "Geles", "Diandra", "Jada", "Alexis", "Nurys", "Zoe", "Alivia", "Uche", "Audrey", "Nicole"], 4)
# week7 = (["Keyana", "Geles", "Diandra", "Zoe", "Uche", "Nurys", "Jada", "Alexis", "Audrey", "Alivia", "Nicole"], 5)
# week8 = (["Alivia", "Geles", "Diandra", "Alexis", "Jada", "Nurys", "Audrey", "Uche", "Keyana", "Zoe", "Nicole"], 3)


#Semana que deu apagão
currentWeek = []#"Cami", "Tori", "Mik", "Nic", "Kay", "Vic", "Fran", "Alyssa", "Julia", "Emma"]

#Atualizar toda semana
# lista das semanas que já 
allWeeks = [week1] #, week2, week3, week4, week5, week6, week7, week8

# the matches that got denied in the truth booth ("guy name", "girl name")
truthBooth_denied = [("Ethan", "Keyana"), ] #("Anthony", "Geles"), ("Malcom", "Nurys"), ("Dimitri", "Nicole"), ("Clinton", "Uche"), ("Keith", "Alexis"), ("Keith", "Alivia"), ("Michael", "Audrey")

# the matches that were confirmed in the truth booth
truthBooth_confirmed = []#("Tyler", "Nicole")]#("Cameron", "Mik"), ("Sam", "Alyssa")]
```

```python
#retorna o número de correspondências em comum entre duas listas de correspondências
def correlation(list1, list2):
    total = 0
    for i in range(len(list1)):
        if list1[i] == list2[i]:
            total = total + 1
    return total

# retorna se uma lista de correspondências quebra uma regra
def isImpossible(matchlist):
    for match in truthBooth_denied:
        if (matchlist[guys.index(match[0])] == match[1]):
            return True
    for match in truthBooth_confirmed:
        if (matchlist[guys.index(match[0])] != match[1]):
            return True
    for week in allWeeks:
        if correlation(matchlist, week[0]) != week[1]:
            return True        
    return False  
```

```python
#lista de possíveis matches
possible = []

iterable = itertools.permutations(girls,len(girls))
for matching in iterable:
    # skip match lists that break a rule
    if isImpossible(matching):
        continue
    else:
        possible.append(matching)
        
print("Existem " + str(len(possible)) + " possibilidades de match!")

if len(possible) == 1:
  print("Os pares ideais são ", possible )
```

```python

def printRemaining():
    for matching in possible:
        printString = ""
        for i in range(len(matching)):
            printString = printString + ("("+guys[i]+", " + matching[i]+ ")")
        print(printString)
        print("")
```

```python
#Determine quantos casais são iguais entre 2 correspondências
#Oops, acabei de perceber que esta é a mesma função de 'correlação'
def couplesInCommon(matching1, matching2):
    count = 0
    for i in range(0, len(matching1)):
        if matching2[i] == matching1[i]:
            count = count + 1
    return count

def makePercent(n):
    return (float(n) / len(possible))*100

#Determine as chances de um apagão
# Que porcentagem de combinações possíveis restantes contém 0 dos casais desta semana?
beamDistribution = [0] * (len(girls)+1)
for matching in possible:
    commonCount = couplesInCommon(matching, currentWeek)
    beamDistribution[commonCount] = beamDistribution[commonCount] + 1

mostLikelyBeamCount = 0
maxBeamCount = 0
highestBeamIndex = 0
lowestBeamIndex = 0
for i in range(0, len(beamDistribution)):
    beamCount = beamDistribution[i]
    if beamCount != 0:
        highestBeamIndex = i
    if beamCount > maxBeamCount:
        maxBeamCount = beamCount
        mostLikelyBeamCount = i
for i in range(0, len(beamDistribution)-1):
    beamCount = beamDistribution[i]
    if beamCount != 0:
        lowestBeamIndex = i
        break
```

```python
#exclui correspondências perfeitas da cabine da verdade
blackOutPercent = makePercent(beamDistribution[len(truthBooth_confirmed)])
blackOutPercent
```

```python
#Porcentagens de Distribuição de Feixe
distString = ''
for i in range(0, len(beamDistribution)):
    distString = distString + ', '+ str(makePercent(beamDistribution[i]))[:str(makePercent(beamDistribution[i])).index('.')]+'%'

#Imprimir dados de feixe
print('Probabilidades de feixe: [' + distString[2:] + ']')
print('Contagem de feixe mais provável: ' + str(mostLikelyBeamCount))
print('Contagem máxima de feixe: ' + str(highestBeamIndex))
print('Probabilidades de apagão: ' + str(blackOutPercent)+'%')
```

```python
# inicializa o dicionário
match_dictionary = {}
for guy in guys:
    match_dictionary[guy] = [0] * len(girls)

#preenche o dicionário
for matching in possible:
    for guy in guys:
        match_dictionary[guy][girls.index(matching[guys.index(guy)])] += float(1)/len(possible)
```

```python
#determinar a melhor correspondência
def determineBestMatching():
    best_match_score = 0
    best_matching = []
    for matching in possible:
        score = 0
        for guy in guys:
            score += match_dictionary[guy][girls.index(matching[guys.index(guy)])]
        #print( "score " + str(score))
        if score > best_match_score:
            best_match_score = score
            best_matching = matching
    print("Best score " + str(best_match_score))

    best_string = ""
    for i in range(len(best_matching)):
        best_string += guys[i] + ", " +best_matching[i]+"""'''"""
    return best_string
```
