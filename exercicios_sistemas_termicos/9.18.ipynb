```python
!pip install pyromat
import pyromat as pm

pm.config["unit_pressure"] = "kPa"
pm.config["def_p"] = 100

mp_water = pm.get("mp.H2O")
```

# Exercício 9.18

```python
kelvin = 273.15
T1 = 45 + kelvin
```

```python
p1 = mp_water.ps(T=T1)                       # Pressão no condensador em kPa
print(f"p1 = {round(float(p1),1)} kPa")

s1 = mp_water.ss(p=p1)[0]
print(f"s1 = {round(float(s1),4)} kJ/kg K")

T2 = 200 + kelvin                      # Temperatura de entrada da caldeira
p2 = 5000              # Pressão na caldeira em kPa
print(f"p2 = {round(float(p2),1)} kPa")

s2 = s1
v = 1/mp_water.ds(p=p1)[0]
print(f'v1 = {round(float(v),6)} m³/kg')

w_p = v*(p2-p1)
print(f"Trabalho requerido na bomba: {round(float(w_p),2)} kJ/kg")

h1 = mp_water.hs(p=p1)[0]

print(f"h1 = {round(float(h1),2)} kJ/kg")

h2 = h1+w_p

print(f"h2 = {round(float(h2),2)} kJ/kg")

p3 = 5000
T3 = 900 + kelvin                            # Temperatura de saida da caldeira
s3 = mp_water.s(p=p3, T=T3)[0]
print(f"s3 = {round(float(s3),4)} kJ/kg")

p4 = 1000
h4 = mp_water.hs(p=p4)[0]
print(f"h4 = {round(float(h4),2)} kJ/kg")

s4 = mp_water.ss(p=p4)[0]
print(f"s4 = {round(float(s4),4)} kJ/kg")

v4 = 1/mp_water.ds(p=p4)[0]
print(f'v4 = {round(float(v4),6)} m³/kg')

p7 = p3
w_p2 = v4*(p7-p4)

print(f"Trabalho requerido na bomba 2: {round(float(w_p2),3)} kJ/kg")

h7 = h4 + w_p2
print(f"h7 = {round(float(h7),2)} kJ/kg")
```
