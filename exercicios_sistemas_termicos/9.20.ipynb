```python
import pyromat as pm
import numpy as np

pm.config["unit_pressure"] = "kPa"
pm.config["def_p"] = 100

mp_water = pm.get("mp.H2O") # <-- for multi-phase water properties
```

# Exercício 9.20

```python
p1 = 10 # <-- saida do condensador
p2 = 3000 # <-- saida da bomba
p3 = 800 # <-- saida do gerador
p4 = 3500 # <-- entrada da turbina/saida da caldeira
p5 = p4 # <-- saida da turbina
T4 = 400 + 273.15 # K <-- saida da turbina
p6 = p1 # <-- entrada no condensador/saida da turbina
```

```python
h1 = mp_water.hs(p=p1)[0]
print(f"h1 = {round(float(h1),4)}")

v1 = 1/mp_water.ds(p=p1)[0]
print(f"v1 = {round(float(v1),6)}")

w_pump1 = v1*(p2-p1)

print(f"Trabalho gerado na bomba 1: {round(float(w_pump1),5)} kJ/kg")

h2 = h1+w_pump1
print(f"h2 = {round(float(h2),4)}")

h4 = mp_water.h(p=p4, T=T4)
print(f"h4 = {round(float(h4),4)}")

h5 = mp_water.hs(p=p5)[1]
s5 = mp_water.ss(p=p5)[1]
print(f"h5 = {round(float(h5),4)}")
print(f"s5 = {round(float(s5),4)}")

s6 = s5
print(f"s6 = {round(float(s6),4)} kJ/kg K")

T6, x6 = mp_water.T_s(s=s6, p=p6, quality=True)
h6 = mp_water.h(x=x6, p=p6)
print(f"h6 = {round(float(h6),4)}")

print(f"Qualidade do vapor de água: {round(float(x6),4)}")

h3 = mp_water.hs(p=p3)[0]
print(f"h3 = {round(float(h3),4)}")

hf800 = h3
x = (h3 - h2) / (h5 - hf800)
print(f'x = {round(float(x),4)}')
w_t = h4 - h5 + (1 - x)*(h5 - h6)
print(f"Trabalho gerado na turbina: {round(float(w_t),5)} kJ/kg")

q_H = h4-h3
print(f"Calor gerado pela caldeira: {round(float(q_H),1)} kJ/kg")
```

```python
w_net = w_t - w_pump1
print(f"wnet é: {round(float(w_net),2)} kJ/kg")
```

Podemos agora calcular a eficiência térmica do sistema com: $$\eta_{th}=\frac{w_{net}}{q_H}$$

```python
eta_th = (w_net / q_H)
print(f"A eficiência térmica é: {round(float(eta_th),2)*100}%")
```
