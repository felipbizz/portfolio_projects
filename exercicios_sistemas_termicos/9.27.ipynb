```python
kelvin = 273.15
p1 = 3500
T1 = 400 + kelvin
p4 = 10
T4 = 35 + kelvin
p5 = 3700
p6 = 3600
T6 = 30 + kelvin
```

```python
h1 = mp_water.h(T=T1, p=p1)
s1 = mp_water.s(T=T1, p=p1)

print(f"h1 = {round(float(h1),2)} kJ/kg")
print(f"s1 = {round(float(s1),2)} kJ/kg K")

v = 1/mp_water.ds(p=p1)[0]
print(f"Volume específico: {round(float(v),5)} m^3/kg")

h2 = 3165.7
s2 = 6.7675

s3s = s2
x3s = 0.8157
h3s = 2143.5

w_t_s = h2 - h3s
print(f"O trabalho gerado na turbina é: {round(float(w_t_s),2)} kJ/kg")

n1 = 0.88

w_t_ac = n1 * w_t_s
print(f"W_t_ac: {round(float(w_t_ac),2)} kJ/kg")

h3 = h2 - w_t_ac
print(f"h3 = {round(float(h3),2)} kJ/kg")

vf = 0.001006
w_p_s = vf*(p5-p4)
print(f"w_p_s: {round(float(w_p_s),2)} kJ/kg")

n2 = 0.8
w_p_ac = w_p_s/n2
print(f"w_p_ac: {round(float(w_p_ac),2)} kJ/kg")

h6 = mp_water.h(T=T6, p=p6)
print(f"h6 = {round(float(h6),2)} kJ/kg")

qH = h1 - h6
print(f"qH: {round(float(qH),2)} kJ/kg")

eta_th = w_t_ac/qH*100
print(f"A eficiência térmica é: {round(float(eta_th),3)}%")
```
