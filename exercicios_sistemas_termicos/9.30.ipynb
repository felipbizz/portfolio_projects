```python
!pip install pyromat
import pyromat as pm

pm.config["unit_pressure"] = "kPa"
pm.config["def_p"] = 100

mp_water = pm.get("mp.H2O")
```

## Resolução 9.30

```python
kelvin = 273.15
T1 = 30 + kelvin
p2 = 5000
p3 = 5000
T3 = 450 + kelvin
p5 = 500
T5 = T3
p4 = p5
T6 = T1
n_turbina = 0.87
n_bomba = 0.82
pot_total = 10_000
```

```python
p1 = mp_water.p(T=T1)
print(f"p1 = {round(float(p1),1)} kPa")

s1 = mp_water.ss(p=p1)[0]
v = 1/mp_water.ds(p=p1)[0]
print(f"Volume específico: {round(float(v),6)} m³/kg")

h2 = mp_water.h(p=p2)[0]
print(f"h2 = {round(float(h2),1)} kJ/kg")
```

```python
s3 = mp_water.s(T=T3, p=p3)
s3 = 6.8185
print(f"s3 = {round(float(s3),4)} kJ/kg K")

T3, x3 = mp_water.T_s(s=s3,p=p3, quality=True)
print(f"T3 = {round(float(T3),1)} ºC")
print(f"Qualidade do vapor x3: {round(float(x3),4)}")

h3 = mp_water.h(T=T3, p=p3)[0]
print(f"h3 = {round(float(h3),1)} kJ/kg")

h4 = mp_water.hs(p=p4)[1]
print(f"h4 = {round(float(h4),1)} kJ/kg")

w_t1 = (h3 - h4) * n_turbina
print(f"Trabalho gerado pela turbina 1: {round(float(w_t1),1)} kJ/kg")
```

```python
h5 = mp_water.h(T=T5, p=p5)[0]
print(f"h5 = {round(float(h5),1)} kJ/kg")

p6 = mp_water.ps(T=T6)
print(p6)
h6s = 2466.3

w_t2 = (h5 - h6s) * n_turbina
print(f"Trabalho gerado pela turbina 2: {round(float(w_t2),1)} kJ/kg")

m = pot_total/(w_t1+w_t2)
print(f'm = {round(float(m),1)} kg/s')
```

```python
w_p = (v*(p3-p6))/n_bomba
print(f"Trabalho requerido pela bomba: {round(float(w_p),1)} kJ/kg")

W_p = w_p*m
print(f"Potencia necessária na bomba: {round(float(W_p),1)} kW")
```

```python
q_H = (h3-h2)+(h5-h4)
print(f"Calor gerado na caldeira: {round(float(q_H),1)} kJ/kg")
```

```python
eta_th = ((w_t1+w_t2)-w_p)/q_H*100
print(f"A eficiência térmica é: {round(float(eta_th),1)}%")
```
