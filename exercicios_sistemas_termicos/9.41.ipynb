```python
pr = 14
p1 = 100
T1 = 300
p2 = pr * p1
p3 = p2
T3 = 1600
k = 1.4
kzin = (k-1)/k
cpo = 1.004
Wnet = 100_000
n_compressor = 0.85
n_turbina = 0.88
```

```python
T2 = T1*(pr)**kzin
print(f"T2 = {round(float(T2),1)} K")

wc = cpo*(T2-T1)
print(f"O trabalho no compressor é: {round(float(wc),3)} kJ/kg")

p4 = p1
T4 = T3*(p4/p3)**kzin
print(f"T4 = {round(float(T4),1)} K")

wt = cpo*(T3-T4)
print(f"O trabalho na turbina é: {round(float(wt),3)} kJ/kg")

wnet = wt - wc
print(f"wnet = {round(float(wnet),3)} kJ/kg")

m = Wnet/wnet
print(f"m = {round(float(m),3)} kg/s")

Wt = m*wt
print(f"Wt = {round(float(Wt),3)} MW")

n1 = (wc/wt)*100
print(f"A eficiência 1: {round(float(n1),1)}%")
```

```python
Tx = (0.75 * (T4-T2)) + T2
print(f"Tx = {round(float(Tx),1)} K")

qh = cpo*(T3-Tx)
print(f"qh = {round(float(qh),3)} kJ/kg")

n = (wnet / qh)*100
print(f"A eficiência térmica é: {round(float(n),1)}%")
```
