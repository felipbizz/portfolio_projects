# Exercício 9.43
Um ciclo turbina a gas usado para uso veiuclar
Na primeira turbina o gás expande expande até uma pressão p5, suficiente para que a turbina acione o compressor. O gas então expandido em uma segunda turbina que aciona as rodas motrizes. Considerando que o fluxo de trabalho seja ar através de todo o ciclo, determine: pressão intermediária, trabalho liquido desenvolvido pelo motor, temperatura do ar na seção de entrada da camara e etc. 

```python
!pip install pyromat
import pyromat as pm

pm.config["unit_pressure"] = "kPa"
pm.config["def_p"] = 100

mp_water = pm.get("mp.H2O")
```

```python
p1 = 100                             # Pressão no condensador em kPa
T1 = 300
k = 1.4
p2 = 600                           # Pressão na caldeira em kPa
cpo = 1.0035
T4 = 1600
p4 = p2
p6 = p7 = 100
```

```python
print(f"T1 = {round(float(T1),1)} K")

psi = (k-1)/k

s1 = mp_water.ss(p=p1)[0]

s2 = s1
v = 1/mp_water.ds(p=p1)[0]

w_p = v*(p2-p1)
print(f"Trabalho requerido na bomba: {round(float(w_p),1)} kJ/kg")

h1 = mp_water.hs(p=p1)[0]

print(f"h1 = {round(float(h1),1)} kJ/kg")

h2 = h1+w_p

T2s = T1*((p2/p1) ** psi) 
print(f"T2s = {round(float(T2s),1)} ºC")

print(f"h2 = {round(float(h2),1)} kJ/kg")

w_c = cpo * (T1-T2s)
print(f"w_c = {round(float(w_c),1)} ")

T5 = T4 - abs(w_c)/cpo
print(f"T5 = {round(float(T5),1)} ºC")

p5s = p4 / ((T4/T5)**(1/psi))
print(f"p5s = {round(float(p5s),1)} kPa")

T6s = T5 / ((p5s/p6) **(psi))
print(f"T6s = {round(float(T6s),1)} ºC")
```

O fluxo de massa é dado por:

```python
T3 = T6s
T7 = T2s

W_t2 = 150
w_t2 = cpo * (T5-T6s)
mdot = W_t2/w_t2
print(f"mdot = {round(float(mdot),3)} kJ/kg")
```

Com o fluxo de massa pode-se calcular a potencia da turbina 1

```python
w_t1 = abs(w_c)
W_t1 = w_t1 * mdot
W_t1
```

Considerando que a turbina 1 é utilizada especificamente para suprir o compressor

```python
W_c = -W_t1
W_c
```

Fluxo da calor cedido na camara de combustão é dado por:

```python
Q_cc = mdot * cpo * (T4-T3)
Q_cc
```

```python
Wliquido = W_t2 
n = (Wliquido / Q_cc)*100
print(f"O rendimento térmico é = {round(float(n),4)}")
```

Caso nao seja utilizado regenerador:

```python
Q_queimador = mdot * cpo * (T4-T2s)
Q_queimador
```

```python
new_w_t2 = cpo * (T5 - T6s)
new_w_t2
```

```python
new_n = (Wliquido/Q_queimador)*100
print(f"O novo rendimento térmico é = {round(float(new_n),4)}")
```
