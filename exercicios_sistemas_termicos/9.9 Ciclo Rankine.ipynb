# Enunciado Problema 9.9
Determine a disponibilidade de água nos quatro estados do ciclo de Rankine descrito no problema anterior. Admita que o reservatório de alta temperatura esteja a 500ºC e que o de baixa esteja a 25ºC. Determine os fluxos de disponibilidade nos reservatórios, por quilograma de vapor que escoa no ciclo, e qual é a eficiencia do ciclo baseada na segunda lei da termodinamica.

O problema foi resolvido utilizando a biblioteca PYroMat, desenvolvido em Python. Para começar, são importados os dados de tabela e configurado suas unidades:

```python
!pip install pyromat
import pyromat as pm

pm.config["unit_pressure"] = "kPa"
pm.config["def_p"] = 100

mp_water = pm.get("mp.H2O")
```

Estado de referência:

```python
po = 100                        #  kPa
To = 25 + 273.15                #  °C 
so = 0.3674                     #  kJ/kg K, 
ho = 104.89                     #  kJ/kg
```

Resolução Exercício 9.8

```python
p1 = 10                             # Pressão no condensador em kPa
T1 = mp_water.Ts(p=p1)[0]
print(f"T1 = {round(float(T1),1)} ºC")

s1 = mp_water.ss(p=p1)[0]
p2 = 3500                           # Pressão na caldeira em kPa
s2= s1
v = 1/mp_water.ds(p=p1)[0]

w_p = v*(p2-p1)
print(f"Trabalho requerido na bomba: {round(float(w_p),1)} kJ/kg")

h1 = mp_water.hs(p=p1)[0]

print(f"h1 = {round(float(h1),1)} kJ/kg")

h2 = h1+w_p
T2 = 400                            # Temperatura na saída da caldeira em ºC

print(f"h2 = {round(float(h2),1)} kJ/kg")

p3 = p2
s3 = mp_water.ss(p=p3)[1]
T3, x3 = mp_water.T_s(s=s3,p=p3, quality=True)
s3dash = mp_water.ss(p=p3)[0]
T3dash = T3

print(f"T3 = {round(float(T3),1)} ºC")
print(f"Qualidade do vapor x3: {round(float(x3),4)}")

h3 = mp_water.hs(p=p3)[1]

print(f"h3 = {round(float(h3),1)} kJ/kg")

print(f"s3 = {round(float(s3),1)} kJ/kg K")

q_H = h3-h2

print(f"Calor gerado na caldeira: {round(float(q_H),1)} kJ/kg")

p4 = p1
s4 = s3
T4, x4 = mp_water.T_s(s=s4,p=p4, quality=True)

print(f"T4 = {round(float(T4),1)} ºC")

h4 = mp_water.h(p=p4,x=x4)

print(f"h4 = {round(float(h4),1)} kJ/kg")
print(f"s4 = {round(float(s4),1)} kJ/kg K")

w_t = h3-h4
print(f"Qualidade do vapor de baixa pressão x4: {round(float(x4),4)}")
print(f"Trabalho gerado pela turbina: {round(float(w_t),1)} kJ/kg")

q_L = h4-h1
print(f"Calor rejeitado pelo condensador: {round(float(q_L),1)} kJ/kg")
```

Disponibilidade de água nos 4 estados do ciclo:

```python
psi1 = float(h1) - ho - To * (s1-so)
print(f"Disponibilidade no estado 1 = {round(float(psi1),2)} kJ/Kg")
```

```python
psi2 = h2 - ho - To * (s1-so)
print(f"Disponibilidade no estado 2 = {round(float(psi2),2)} kJ/Kg")
```

```python
psi3 = h3 - ho - To * (s3-so)
print(f"Disponibilidade no estado 3 = {round(float(psi3),2)} kJ/Kg")
```

```python
psi4 = h4 - ho - To * (s4-so)
print(f"Disponibilidade no estado 4 = {round(float(psi4),2)} kJ/Kg")
```

```python
Th = 500      #Reservatório de alta temperatura 
```

```python
delta_psi_h = abs(((1-To)/Th)*q_H)
print(f"{round(float(delta_psi_h),2)} kJ/Kg")
```

```python
delta_psi_L = (1-To/To) * q_L
print(f"{round(float(delta_psi_L),2)} kJ/Kg")
```

```python
n = (w_t - w_p) / delta_psi_h
print(f"A eficiência do ciclo é de: {round(float(n*100),2)}%")
```
